```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
bg = cv2.imread("josh-berquist-0gNZxc5ds58.jpg")
bg = cv2.cvtColor(bg, cv2.COLOR_BGR2RGB)
plt.imshow(bg)
```

```python
watermark = cv2.imread("pngwing lamborghini.png", -1)
wh, ww, wc = watermark.shape


# plt.imshow(watermark[watermark==255])
print(watermark.shape)
```

```python
(h, w, c) = bg.shape
image = np.dstack([bg, np.full((h, w), 255)])
print(image.shape)
plt.imshow(image)
```

```python
x1, y1 = 20, 20
x2, y2 = x1 + ww, y1 + wh

overlay = np.zeros_like(image)
overlay[y1:y2, x1:x2] = watermark
```

```python
plt.imshow(overlay)
```

```python
combined = image.copy()
combined = cv2.addWeighted(overlay, 0.9, combined, 1.0, 0.0)

plt.imsave("watermark_img.png", combined.astype("uint8"))
plt.imshow(combined)
```
